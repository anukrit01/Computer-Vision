```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import cv2
```

```python
# Get the image with imread
img = cv2.imread("desktop/pic.jpg")
```

```python
# Check the type
type(img)
```

```python
# Shape
img.shape
```

```python
plt.imshow(img)
```

```python
# Convert OpenCV to the channels of photo
img_fix = cv2.cvtColor(img,
                      cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_fix)
```

```python
# Scale it to gray and check the shape
img_gray = cv2.imread('desktop/pic.jpg',
                     cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap= 'gray')
```

```python
# Resize the image
img_new = cv2.resize(img_fix,
                    (1000, 400))
```

```python
plt.imshow(img_new)
```

```python
# Resize with ratio
width_ratio = 0.5
height_ratio = 0.5
```

```python
img2 = cv2.resize(img_fix,
                 (0, 0),
                 img_fix,
                 width_ratio,
                 height_ratio)
```

```python
plt.imshow(img2)
```

```python
img2.shape
```

```python
# Flip on horizontal axis
img3 = cv2.flip(img_fix, 0)
plt.imshow(img3)
```

```python
# Flip on vertical axis
img3 = cv2.flip(img_fix, 1)
plt.imshow(img3)
```

```python
# Flip on vertical & horizontal axis
img3 = cv2.flip(img_fix, -1)
plt.imshow(img3)
```

```python
# Change the size of canvas
last_img = plt.figure(figsize=(10, 7))
ilp = last_img.add_subplot(111)
ilp.imshow(img_fix)
```
